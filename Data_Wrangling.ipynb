#Data Normalization


1.  Simple feature scaling
2.  min-max


1.   Z-score






```python
from os import WIFEXITED
mydata['width'].head()
```

```python
mydata[['width', 'height', 'length']].head()
```

```python
# data normalization: simple feature scaling
mydata['width'] = mydata['width']/ mydata['width'].max()

mydata[['width', 'height', 'length']].head()
```

```python
# min-max
mydata['length'] = (mydata['length'] - mydata['length'].min()) / (mydata['length'].max() - mydata['length'].min())

mydata[['width', 'height', 'length']].head()
```

```python
# Z-score
mydata['height'] = (mydata['height'] - mydata['height'].mean()) / (mydata['height'].std())

mydata[['width', 'height', 'length']].head()
```

#One hot-encoding
One-Hot Encoding is a technique used to convert categorical data into a format that can be provided to machine learning algorithms to improve predictions. This method creates new binary columns (or features), where each column represents one possible category of the original data.

```python
#one hot-encoding

conv_fuel = pd.get_dummies(mydata['fuel-type'], dtype=int)

conv_fuel.head()
```

```python
mydata_new = pd.concat([mydata, conv_fuel], axis=1)

mydata_new.head()
```
